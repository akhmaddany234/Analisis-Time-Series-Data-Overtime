# Time Series Analysis & Model Building

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller, acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
import pmdarima as pm
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
daily_ts = pd.read_csv('processed_daily_overtime.csv', parse_dates=['Date'], index_col='Date')
print(f"Time Series Range: {daily_ts.index.min()} to {daily_ts.index.max()}")
print(f"Total Days: {len(daily_ts)}")
print("\nFirst 5 rows:")
display(daily_ts.head())
```

Fill any missing dates

```python
ts_data = daily_ts['Total_Overtime_Hours']
```

3. Time Series Decomposition

```python
# Decompose time series (assuming weekly seasonality)
decomposition = seasonal_decompose(ts_data, model='additive', period=7)

fig, axes = plt.subplots(4, 1, figsize=(15, 10))

# Original
axes[0].plot(ts_data, label='Original', color='blue')
axes[0].set_ylabel('Overtime Hours')
axes[0].set_title('Time Series Decomposition of Overtime Hours', fontsize=16, fontweight='bold')
axes[0].legend()

# Trend
axes[1].plot(decomposition.trend, label='Trend', color='red')
axes[1].set_ylabel('Trend')
axes[1].legend()

# Seasonal
axes[2].plot(decomposition.seasonal, label='Seasonal', color='green')
axes[2].set_ylabel('Seasonal')
axes[2].legend()

# Residual
axes[3].plot(decomposition.resid, label='Residual', color='orange')
axes[3].set_ylabel('Residual')
axes[3].set_xlabel('Date')
axes[3].legend()

plt.tight_layout()
plt.show()
```

4. Stationarity Test (Augmented Dickey-Fuller)

```python
def test_stationarity(timeseries, window=7):
    """Test stationarity of time series"""
    
    # Rolling statistics
    rolmean = timeseries.rolling(window=window).mean()
    rolstd = timeseries.rolling(window=window).std()
    
    # Plot rolling statistics
    fig, axes = plt.subplots(2, 1, figsize=(12, 8))
    
    axes[0].plot(timeseries, color='blue', label='Original')
    axes[0].plot(rolmean, color='red', label='Rolling Mean')
    axes[0].plot(rolstd, color='black', label='Rolling Std')
    axes[0].set_title('Rolling Mean & Standard Deviation', fontsize=14)
    axes[0].legend()

        # ADF Test
    axes[1].plot(timeseries.diff().dropna(), color='green')
    axes[1].set_title('First Difference', fontsize=14)
    axes[1].set_xlabel('Date')
    
    plt.tight_layout()
    plt.show()
    
    # Perform Dickey-Fuller test
    print('Results of Dickey-Fuller Test:')
    adftest = adfuller(timeseries.dropna(), autolag='AIC')
    adfoutput = pd.Series(adftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    
    for key, value in adftest[4].items():
        adfoutput[f'Critical Value ({key})'] = value
    
    print(adfoutput)

    if adfoutput['p-value'] < 0.05:
        print("\nConclusion: Series is STATIONARY")
    else:
        print("\nConclusion: Series is NON-STATIONARY")
    
    return adfoutput

# Test stationarity
adf_result = test_stationarity(ts_data)
```

5. Differencing to Achieve Stationarity

```python
# Apply differencing if not stationary
if adf_result['p-value'] >= 0.05:
    print("Applying differencing to achieve stationarity...")
    ts_diff = ts_data.diff().dropna()
    
    # Test stationarity again
    adf_result_diff = test_stationarity(ts_diff)
else:
    ts_diff = ts_data
    print("Series is already stationary, no differencing needed.")
```

6. ACF and PACF Plots

```python
# %%
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Original series
plot_acf(ts_data, lags=20, ax=axes[0, 0])
axes[0, 0].set_title('ACF - Original Series')

plot_pacf(ts_data, lags=20, ax=axes[0, 1])
axes[0, 1].set_title('PACF - Original Series')

# Differenced series (if applied)
if 'ts_diff' in locals():
    plot_acf(ts_diff, lags=20, ax=axes[1, 0])
    axes[1, 0].set_title('ACF - Differenced Series')
    
    plot_pacf(ts_diff, lags=20, ax=axes[1, 1])
    axes[1, 1].set_title('PACF - Differenced Series')

plt.tight_layout()
plt.show()
```

8. ARIMA Model

```python
def evaluate_model(y_true, y_pred, model_name):
    """Calculate evaluation metrics"""
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mape = mean_absolute_percentage_error(y_true, y_pred) * 100
    
    metrics = {
        'MAE': mae,
        'MSE': mse,
        'RMSE': rmse,
        'MAPE': mape
    }
    
    print(f"\n{model_name} Evaluation:")
    print(f"MAE: {mae:.2f}")
    print(f"MSE: {mse:.2f}")
    print(f"RMSE: {rmse:.2f}")
    print(f"MAPE: {mape:.2f}%")

    return metrics
```

```python
# Fit ARIMA model
print("Fitting ARIMA model...")
try:
    # Using auto_arima to find best parameters
    auto_model = pm.auto_arima(train, 
                              seasonal=False, 
                              trace=True,
                              error_action='ignore',
                              suppress_warnings=True,
                              stepwise=True)
    
    print(f"\nBest ARIMA parameters: {auto_model.order}")
    
    # Fit model with best parameters
    arima_model = ARIMA(train, order=auto_model.order)
    arima_fit = arima_model.fit()
    
    print("\nARIMA Model Summary:")
    print(arima_fit.summary())
    
    # Make predictions
    arima_pred = arima_fit.forecast(steps=len(test))
    
    # Evaluate
    arima_metrics = evaluate_model(test, arima_pred, "ARIMA")
    
    # Plot predictions
    plt.figure(figsize=(12, 6))
    plt.plot(train.index, train, label='Train', color='blue')
    plt.plot(test.index, test, label='Test', color='green')
    plt.plot(test.index, arima_pred, label='ARIMA Predictions', color='red', linestyle='--')
    plt.title('ARIMA Model Predictions vs Actual', fontsize=14, fontweight='bold')
    plt.xlabel('Date')
    plt.ylabel('Overtime Hours')
    plt.legend()
    plt.tight_layout()
    plt.show()
    
except Exception as e:
    print(f"Error in ARIMA: {e}")
    # Fallback to simple ARIMA(1,1,1)
    print("Trying with default parameters (1,1,1)...")
    arima_model = ARIMA(train, order=(1,1,1))
    arima_fit = arima_model.fit()
    arima_pred = arima_fit.forecast(steps=len(test))
    arima_metrics = evaluate_model(test, arima_pred, "ARIMA")
```

9. SARIMA Model (with weekly seasonality)

```python
# Fit SARIMA model (weekly seasonality: s=7)
print("\n" + "="*50)
print("Fitting SARIMA model...")
print("="*50)

try:
    # Using auto_arima with seasonal component
    sarima_auto = pm.auto_arima(train,
                               seasonal=True,
                               m=7,  # Weekly seasonality
                               trace=True,
                               error_action='ignore',
                               suppress_warnings=True,
                               stepwise=True)
    
    print(f"\nBest SARIMA parameters: {sarima_auto.order}, Seasonal: {sarima_auto.seasonal_order}")
    # Fit SARIMA model
    sarima_model = SARIMAX(train,
                          order=sarima_auto.order,
                          seasonal_order=sarima_auto.seasonal_order)
    sarima_fit = sarima_model.fit(disp=False)
    
    print("\nSARIMA Model Summary:")
    print(sarima_fit.summary())
    
    # Make predictions
    sarima_pred = sarima_fit.forecast(steps=len(test))
    
    # Evaluate
    sarima_metrics = evaluate_model(test, sarima_pred, "SARIMA")
    # Plot predictions
    plt.figure(figsize=(12, 6))
    plt.plot(train.index, train, label='Train', color='blue')
    plt.plot(test.index, test, label='Test', color='green')
    plt.plot(test.index, sarima_pred, label='SARIMA Predictions', color='red', linestyle='--')
    plt.title('SARIMA Model Predictions vs Actual', fontsize=14, fontweight='bold')
    plt.xlabel('Date')
    plt.ylabel('Overtime Hours')
    plt.legend()
    plt.tight_layout()
    plt.show()
    
    # Plot diagnostics
    sarima_fit.plot_diagnostics(figsize=(12, 8))
    plt.suptitle('SARIMA Model Diagnostics', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()
except Exception as e:
    print(f"Error in SARIMA: {e}")
    print("SARIMA model failed, using simpler approach...")
```

### Uji Stasioneritas (Augmented Dickey-Fuller Test)

Sebelum memodelkan deret waktu (seperti ARIMA), kita harus menguji apakah data kita stasioner (rata-rata dan varians stabil). Kita menggunakan uji Augmented Dickey-Fuller (ADF).

```python
from statsmodels.tsa.stattools import adfuller

print("Hasil Uji Augmented Dickey-Fuller (ADF):")
adf_result = adfuller(ts_data)
print(f"ADF Statistic: {adf_result[0]:.4f}")
print(f"p-value: {adf_result[1]:.4f}")
print("Critical Values:")
for key, value in adf_result[4].items():
    print(f"  {key}: {value:.4f}")

if adf_result[1] < 0.05:
    print("\nKesimpulan: p-value < 0.05, sehingga kita menolak hipotesis nol (H0). Data adalah STASIONER.")
else:
    print("\nKesimpulan: p-value >= 0.05, sehingga kita gagal menolak hipotesis nol (H0). Data TIDAK STASIONER.")
```

### Analisis Autokorelasi (ACF dan PACF)

Plot Autocorrelation Function (ACF) dan Partial Autocorrelation Function (PACF) membantu kita melihat korelasi antara observasi saat ini dengan lag (waktu sebelumnya) dan menentukan orde untuk model ARIMA (p, d, q).

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Plot ACF
plot_acf(ts_data, ax=axes[0], lags=40, title='Autocorrelation Function (ACF)')

# Plot PACF
plot_pacf(ts_data, ax=axes[1], lags=40, title='Partial Autocorrelation Function (PACF)')

plt.tight_layout()
plt.show()
```
